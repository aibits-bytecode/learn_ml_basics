L1 and L2 Regulerization

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('data/Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance',
               'CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price'
              ]
dataset = dataset[cols_to_use]
dataset.head()
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount','Distance','Bedroom2','Bathroom','Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
```

```python
dataset.isna().sum()
```

```python
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.shape
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
dataset.head()
```

```python
x = dataset.drop('Price', axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
reg.score(x_train,y_train)
```

L1 regulerization

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_test, y_test)
```

```python
lasso_reg.score(x_train, y_train)
```

L2 regularization

```python
from sklearn.linear_model import Ridge
ridge_reg =Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```
