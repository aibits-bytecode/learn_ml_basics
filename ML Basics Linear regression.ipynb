## Home Price Prediction using Linear regression

```python
import pandas as pd
df = pd.read_csv('data/homeprices.csv')
df
```

```python
# In here area is the independent variable
# price is the dependent variable
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df.head()
```

```python
plt.xlabel('area(sqr ft)')
plt.ylabel('price($)')
plt.scatter(df.area, df.price, marker='*', color='r')
```

### Creating the model

```python
linear_reg = linear_model.LinearRegression()
```

### Training the model

```python
linear_reg.fit(df[['area']],df.price)
```

```python
plt.xlabel('area(sqr ft)')
plt.ylabel('price($)')
plt.scatter(df.area, df.price, marker='*', color='r')
plt.plot(df.area, linear_reg.predict(df[['area']]))
```

### Prediction

```python
linear_reg.predict([[3000]])
```

```python
# this gives the gradient of our model function
linear_reg.coef_
```

```python
# this gives the intercept of our model function
linear_reg.intercept_
```

```python
linear_reg.coef_*3000+linear_reg.intercept_
```
